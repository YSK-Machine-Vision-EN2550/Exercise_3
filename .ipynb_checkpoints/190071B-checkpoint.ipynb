```python
########################################################
#ex1
#Box Kernal
########################################################
%matplotlib inline
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator

img = cv.imread('butterfly.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None

kernal_width = 9
sigma = 4
kernal = np.ones((kernal_width,kernal_width),np.float32)/kernal_width**2
gausiankernal = cv.getGaussianKernel(9,sigma)

imgWithKernal = cv.filter2D(img,-1,kernal)
imgWithGausianKernal = cv.filter2D(img,-1,kernal)
    
convertedImg1 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
convertedImg2 = cv.cvtColor(imgWithKernal, cv.COLOR_BGR2RGB)
convertedImg3 = cv.cvtColor(imgWithGausianKernal, cv.COLOR_BGR2RGB)
fig, ax = plt.subplots(3,1, figsize = (20,20))
ax[0].imshow(convertedImg1)
ax[1].imshow(convertedImg2)
ax[2].imshow(convertedImg3)
ax[0].set_title('Original')
ax[1].set_title("Afetr apply the box  Kernal")
ax[2].set_title("Afetr apply the Gausian  Kernal")

plt.show()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
from matplotlib.ticker import FormatStrFormatter
x = np.arange(-5,5.01,0.01)
y = np.arange(-5,5.01,0.01)
x,y = np.meshgrid(x,y)
sigma = 1
GausianKernal = (1/(2*np.pi*sigma**2))*np.exp(-(x**2 + y**2)/(2*sigma**2))


fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111,projection = '3d')
surf = ax.plot_surface(x,y,GausianKernal,cmap =cm.jet, linewidth = 0,antialiased = True)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
csetn= ax.contourf(x,y,GausianKernal, zdir = 'z', offset = np.min(GausianKernal)-1.5,cmap =cm.jet)
ax.set_zlim(np.min(GausianKernal)-2, np.max(GausianKernal))
plt.axis('off')
plt.show()
```
